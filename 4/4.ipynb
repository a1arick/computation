## Нужные функции

```python
# округление 
def o(x):
    return round(x * pow(10,5))/pow(10,5) 

# значение функции 
def getValue(x):
    return math.exp(2*x)

# точное значение певрой производной
def getExpFirstDerivative (x):
    return 2 * math.exp(2*x)

def getExpSecondDerivative (x):
    return 4 *  math.exp(2*x)


# значение первой производной с первый порядком точности 
def getFirstDevivativeOH(x, h):
    return (o(getValue(x)) - o(getValue(x-h)))/h

# значение первой производной со вторым порядком точности 
def getFirstDerivativeOH2(x, h):
    return ((-3)* o(getValue(x)) + 4* o(getValue(x + h)) - o(getValue(x + 2*h)))/(2*h)

# значение второй производной с первым порядком точности 
def getSecondDerivativeOH2(x,  h):
    return (o(getValue(x+h)) - 2*o(getValue(x)) + o(getValue(x-h)))/(pow(h, 2))

# значение первой производнйо с третьим порядком точности 
def getFirstDerivativeOH3(x,  h):
    return ((-11)*o(getValue(x-3*h)) +  18*o(getValue(x-2*h)) - 9*o(getValue(x-h)) + 2*o(getValue(x)))/(6*h)

```

```python
df_1 = pd.DataFrame(columns = ["x", 
                             "f(x)",
                             "f(x)'",
                              "f_'(x)",
                              "O(h)",
                              "f='(x)",
                              "O(h^2)",
                              "f(x)''",
                              "f_(x)''",
                              "O(h^2)"]) 
```

```python
def f_1(df):
    h = 0.0 # шаг
    n = 0 # количтсво узлов
    begin = 0.0 # начало отрезка 
    h = 0.1
    n = 10
    
    for i in range(n):
        x = begin + h * (i)
        derivf = getExpFirstDerivative(x)
        secondderivf = getExpSecondDerivative(x)
        df = df.append({"x" : x, 
                        "f(x)" : getValue(x),
                        "f(x)'" : derivf,
                        "f_'(x)" : getFirstDevivativeOH(x, h),
                        "O(h)" : math.fabs(getFirstDevivativeOH(x, h) - derivf),
                        "f='(x)" :  getFirstDerivativeOH2(x, h),
                        "O(h^2)" : math.fabs( getFirstDerivativeOH2(x, h) - derivf),
                        "f(x)''" : secondderivf,
                        "f_(x)''" : getSecondDerivativeOH2(x, h),
                        "O(h^2)" : math.fabs(getSecondDerivativeOH2(x, h) - secondderivf)
                       }, ignore_index=True)
        
    return df
```

```python
df_1 = f_1(df_1)
```

### 2) Пользуясь формулой $(6)$ в точке $x = 1$ вычислить разностную производную второго порядка
### аппроксимации функции $e^{2x}$, последовательно уменьшая шаг $h$ (например,вдвое) до тех пор, пока 
### фактическая погрешность не начнет возрастать. Определить $h$ оптимальное экпериментально и 
### теоретически, объяснить полученные результаты.Значения функции округлять до 5-ого знака после
### запятой, т. е. $ε = 5 · 10^{−6}$



```python
df_info = pd.DataFrame(columns = ["h", 
                               "f",
                               "погрешность"]) 
```

```python
# вывод поиска оптимального h для первой производной с первым порядком точности
def f_2(df, n):
    x = 1.0
    h = 0.2
    for i in range(n):
        h = h / 2
        df = df.append({"h" : h, 
                        "f" : getFirstDerivativeOH2(x, h),
                        "погрешность" : getExpFirstDerivative(x) - getFirstDerivativeOH2(x, h)
                       }, ignore_index=True)
    return df    
```

```python
df_2 = f_2(df_info, 6) 
df_3 = f_2(df_info, 5)
df_4 = f_2(df_info, 4)
```
